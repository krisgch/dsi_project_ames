# Import

```python
import pandas as pd
from sklearn import metrics
```

# Load train data

```python
train = pd.read_csv('train.csv')
train
```

# Make model

```python
X = train
y = train['SalePrice']
```

```python
def idiot_model(X):
    return [100]*len(X)
```

# Evaluate model

```python
preds = idiot_model(X)
```

```python
metrics.mean_squared_error(y, preds, squared=False)
```

# Prediction & Submission

```python
## load test data
test = pd.read_csv('test_cleaned.csv')
test
```

```python
preds_test = idiot_model(test)
preds_test[:10]
```

```python
## create dataframe with "Order" and prediction
pred_df = pd.DataFrame({'Order':test['Order'], 'SalePrice':preds_test})
pred_df
```

```python
## save to csv file for submission
## index = False
pred_df.to_csv('submission.csv', index=False)
```
