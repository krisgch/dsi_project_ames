```python
# Import usual library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Modeling library
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.dummy import DummyRegressor # (baseline score)
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Lasso, LassoCV

# Using this code bit only to make the notebook cleaner to read
# through. 
from warnings import simplefilter
from sklearn.exceptions import ConvergenceWarning
simplefilter("ignore", category=ConvergenceWarning)

pd.options.mode.chained_assignment = None  # default='warn'

import itertools as it
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('./datasets/train_cleaned.csv')
```

```python
df.isnull().mean()[df.isnull().mean()>0]
```

```python
df['fireplace_quality'] = df['fireplace_quality'].fillna('NA')
```

```python
# Impute values that represent no basement
df['basement_quality'].fillna('NA',inplace=True)
df['basement_condition'].fillna('NA',inplace=True)
df['basement_exposure'].fillna('NA',inplace=True)
df['basement_fin_type_1'].fillna('NA',inplace=True)
df['basement_fin_type_2'].fillna('NA',inplace=True)
df['basement_fin_sf_1'].fillna(0,inplace=True) #mode
df['basement_fin_sf_2'].fillna(0,inplace=True) #mode
df['basement_unf_sf'].fillna(0,inplace=True) #mode
df['basement_full_bath'].fillna(0,inplace=True)
df['basement_half_bath'].fillna(0,inplace=True)
df['total_basement_sf'].fillna(0,inplace=True)
```

```python
# impute garage type as missing values
df['garage_type'].fillna('NA',inplace=True)
df['garage_year_built'].fillna(df['garage_year_built'].mode()[0],inplace=True)
df['garage_fin'].fillna('NA',inplace=True)
df['garage_quality'].fillna('NA',inplace=True)
df['garage_condition'].fillna('NA',inplace=True)
df['garage_cars'].fillna(0,inplace=True)
df['garage_area'].fillna(0,inplace=True)
```

```python
df.isnull().mean()[df.isnull().mean()>0]
```

Generate some feature engineering columns

```python
df['overall_ground'] = df['overall_quality']*df['ground_living_area']
```

```python
df['overall^2'] = df['overall_quality']**2
```

```python
df['overall_garage_car'] = df['overall_quality']*df['garage_cars']
```

```python
df['overall_year'] = df['overall_quality']*df['year_built']
```

```python
df['overall_lot'] = df['overall_quality']*df['lot_area']
```

```python
# Use LASSO Regression to identify key columns against log_sale_price
# Separate X and y
X = df_num.drop(columns=['sale_price','log_sale_price'])
y = df_num['log_sale_price']
```

```python
# Train Test Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Scale our data.
# Relabeling scaled data as "Z" is common.
sc = StandardScaler()
Z_train = sc.fit_transform(X_train)
Z_test = sc.transform(X_test)
```

```python
# Set up a list of Lasso alphas to check.
l_alphas = np.logspace(-3,0,100)
```

```python
# Cross-validate over our list of Lasso alphas
lasso_cv = LassoCV(alphas=l_alphas)
```

```python
# Fit model using best LASSO alpha!
lasso_cv.fit(Z_train, y_train)
```

```python
lasso_coefs = lasso_cv.coef_
```

```python
lasso_coefs = pd.Series(lasso_coefs, X.columns)
```

```python
plt.figure(figsize=(6,6))
lasso_coefs[lasso_coefs != 0].sort_values(ascending=True).plot.barh();
```

Baseline model from LASSO

```python
# features
features = ['overall_quality', 'ground_living_area', 'year_built',
       'overall_condition', 'garage_cars', 'fireplace', 'year_remodel_add',
       'total_basement_sf', 'basement_full_bath', 'basement_fin_sf_1',
       'lot_area', 'screen_porch', 'full_bath', 'wood_deck_sf',
       '1st_floor_sf']
```

```python
# Recheck correlation with columns from LASSO Regression
X = df_num[features]
y = df_num['sale_price']
```

```python
lr = LinearRegression()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
y_train_log = y_train.apply(np.log)
y_test_log = y_test.apply(np.log)
```

```python
lr.fit(X_train,y_train_log)
```

```python
# predict
y_train_preds = lr.predict(X_train)
y_test_preds = lr.predict(X_test)

# convert exponential back to USD
y_train_preds = np.exp(y_train_preds)
y_test_preds = np.exp(y_test_preds)

# train test score
train_r2 = lr.score(X_train,y_train_log)
test_r2 = lr.score(X_test,y_test_log)

# cross validation score
mean_cross_val = cross_val_score(lr, X_train, y_train_log).mean()

# metrics (comparing y_preds with y)
rmse_train = round(metrics.mean_squared_error(y_train,y_train_preds,squared=False),2)
rmse_test = round(metrics.mean_squared_error(y_test,y_test_preds,squared=False),2)

print(" LASSO log Model ".center(30, "="))
print(f"train r2: {train_r2.round(4)}")
print(f"test_r2: {test_r2.round(4)}")
print(f"mean cross val: {mean_cross_val.round(4)}")
print(f"train rmse: {rmse_train}")
print(f"test rmse: {rmse_test}")
```

```python
scores = pd.DataFrame(columns=['model_name','r2_train','r2_test','cross_val','rmse_train','rmse_test'])

for n,col in enumerate(cat_col):
    X = df[col]
    y = df['sale_price']
    X = pd.get_dummies(columns=[col], data = X, drop_first=True)
    lr = LinearRegression()
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.03, random_state=101)
    
    # y_train_log = y_train.apply(np.log)
    # y_test_log = y_test.apply(np.log)
    
    lr.fit(X_train,y_train)
    #lr.fit(X_train,y_train_log)
    
    # predict
    y_train_preds = lr.predict(X_train)
    y_test_preds = lr.predict(X_test)
    
    # convert exponential back to USD
    # y_train_preds = np.exp(y_train_log)
    # y_test_preds = np.exp(y_test_log)

    # train test score
    train_r2 = lr.score(X_train,y_train)
    test_r2 = lr.score(X_test,y_test)

    # cross validation score
    mean_cross_val = cross_val_score(lr, X_train, y_train).mean()

    # metrics (comparing y_preds with y)
    rmse_train = round(metrics.mean_squared_error(y_train,y_train_preds,squared=False),4)
    rmse_test = round(metrics.mean_squared_error(y_test,y_test_preds,squared=False),4)
    
    scores.loc[n] = [col,train_r2, test_r2, mean_cross_val, rmse_train, rmse_test]

scores.sort_values(by='rmse_test').head(15) # test will be our category
```

```python
scores_3 = pd.DataFrame(columns=['model_name','r2_train','r2_test','cross_val','rmse_train','rmse_test'])

for n,col in enumerate(it.combinations(cat_col,3)):
    
    features = ['overall_ground','overall^2','overall_garage_car','overall_year','overall_lot',
            'overall_quality', 'ground_living_area', 'year_built','overall_condition']
    col = list(col)
    features.extend(col)

    X = df[features]
    y = df['sale_price']
    X = pd.get_dummies(columns=col, data = X, drop_first=True)
    lr = LinearRegression()
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
    
    #y_train_log = y_train.apply(np.log)
    #y_test_log = y_test.apply(np.log)
    
    lr.fit(X_train,y_train)
    #lr.fit(X_train,y_train_log)
    
    # predict
    y_train_preds = lr.predict(X_train)
    y_test_preds = lr.predict(X_test)
    
    # convert exponential back to USD
    #y_train_preds = np.exp(y_train_preds)
    #y_test_preds = np.exp(y_test_preds)

    # train test score
    train_r2 = lr.score(X_train,y_train)
    test_r2 = lr.score(X_test,y_test)
    #train_r2 = lr.score(X_train,y_train_log)
    #test_r2 = lr.score(X_test,y_test_log)

    # cross validation score
    mean_cross_val = cross_val_score(lr, X_train, y_train).mean()

    # metrics (comparing y_preds with y)
    rmse_train = round(metrics.mean_squared_error(y_train,y_train_preds,squared=False),4)
    rmse_test = round(metrics.mean_squared_error(y_test,y_test_preds,squared=False),4)
    
    scores_3.loc[n] = ['&'.join(col),train_r2, test_r2, mean_cross_val, rmse_train, rmse_test]
```
